```python
# Question 1: Reverse the string "Python Programming"
string = "Python programming"
reversed_string = string[::-1]
print("reversed_string is:", reversed_string)
```

```python
#Check if "racecar" is a palindrome
word = "racecar"
if word == word[::-1]:
    print(f"{word} is a palindrome.")
```

```python
# Question 7: Check if two strings are anagrams: "listen" and "silent"
s1 = "listen"
s2 = "silent"

if sorted(s1) == sorted(s2):
    print("They are anagrams")
else:
    print("They are not anagrams")
```

```python
# Question 8: Remove all vowels from "Computer Science"
text = "computer science"
vowels = "aeiouAEIOU"
for vowel in vowels:
    text = text.replace(vowel, "")
print("text without vowels is:", text)
```

```python
# Question 9: Remove all punctuation from "Hello, World! How are you?"
import string
text = "hello,world! how are you?"
result = "".join(char for char in text if char not in string.punctuation)
print(result)
```
