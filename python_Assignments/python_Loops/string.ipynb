```python
# Question 1: Reverse the string "Python Programming"
string = "Python programming"
reversed_string = string[::-1]
print("reversed_string is:", reversed_string)
```

```python
#Check if "racecar" is a palindrome
word = "racecar"
if word == word[::-1]:
    print(f"{word} is a palindrome.")
```

```python
# Question 3: Count the number of words in "Python is a great programming language"
sent = "Python is a great programming language"
words = sent.split()
word_count = len(words)
print("the no of words is:", word_count)
```

```python
# Question 7: Check if two strings are anagrams: "listen" and "silent"
s1 = "listen"
s2 = "silent"

if sorted(s1) == sorted(s2):
    print("They are anagrams")
else:
    print("They are not anagrams")
```
