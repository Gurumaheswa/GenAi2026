```python
# question 1: Simple Age Check Write a program that checks if a person can get a driver's license (age >= 16).

age = int(input("Enter your age: "))
if age >= 16:
    print("Eligible to get license")
else :
    print("Under age")
```

```python
# question 2: Positive or Negative Write a program that checks if a number is positive, negative, or zero.

n = int(input("Enter number: "))
if n<0:
    print(f"{n} is a Negative Number")
elif n>0:
    print(f"{n} is a Positive Number")
else:
    print(f"{n} is a Zero")
```

```python
# question 3 :  Larger Number Write a program that compares two numbers and prints which one is larger.

num1 = int(input("Enter number: ")) 
num2 = int(input("Enter number: "))
if num1>num2:
    print(f"{num1} is Greater than {num2}")
elif num1<num2:
    print(f"{num2} is Greater than {num1}")
else:
    print("Both are equal")
```

```python
# question 4 :Leap Year Checker Write a program to check if a year is a leap year:
# Divisible by 4 AND (not divisible by 100 OR divisible by 400)

year = int(input("Enter year: "))
if year%4==0 and (year%100!=0 or year%400==0):
    print(f"{year} is a Leap year")
else:
    print(f"{year} is not a Leap year")
```

```python

#Grade Calculator with Bonus Calculate grades (A-F) but add +1 bonus grade level if attendance > 90%.

Attendence = int(input("Enter attendence percentage: "))
grade = input("Enter your grade:(A-F)").upper()
if Attendence > 90:
    if grade == "B":
        grade = "A"
    elif grade == "C":
        grade = "B"
    elif grade == "D":
        grade = "C"
    elif grade == "E":
        grade = "D"
    elif grade =="F":
        grade = "E"
print(f"Attendence percentage is {Attendence}")
print(f"Final grade is {grade}")
```
