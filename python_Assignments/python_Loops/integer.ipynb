```python
# Question 1: Calculate the product of first 10 natural numbers
product = 1
for i in range(1, 11):
    product *= i

print("The product of the first 10 natural numbers is:", product)


```

```python
# Question 2: Calculate the sum of squares of first 10 natural numbers
sum_of_squares= 0
for i in range(1,11):
    sum_of_squares += i**2
    print("the sum of squares of first 10 natural numbers is:", sum_of_squares)
```

```python
#Calculate the sum of digits in number 12345
num = 12345
sum_of_digits = 0

while num > 0:
    sum_of_digits += num % 10
    num //= 10

print("Sum of digits:", sum_of_digits)

```

```python
#Check if 97 is a prime number
num = 97
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

print("Prime" if count == 2 else "Not Prime")
```

```python
#Find the factorial of 8
num = 8
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial of", num, "is", fact)


```

```python
#Calculate the average of numbers: 15, 23, 31, 42, 56
numbers =[15, 23, 31, 42, 56]
total = sum(numbers)
average = total / len(numbers)
print("Average of the numbers is:", average)
```
